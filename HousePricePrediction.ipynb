# Missing Values
Since there are many NAN values, we must change the categorical values to None and their sub columns to 0.
This is because if lets say a house doesnt have a basement that means their basement's measurements (e.g. area) are 0

```python
df[['MasVnrType', 'FireplaceQu', 'LotFrontage', 'GarageFinish', 'GarageQual','GarageCond', 'GarageType', 'GarageYrBlt', 'BsmtExposure', 'BsmtFinType2','BsmtCond', 'BsmtFinType1', 'BsmtQual', 'MasVnrArea', 'Electrical']].dtypes
```

```python
# since 3 of them are float types, we'll fill 2 with 0s and only the LotFrontage with its median
#This is because there just cant be 0 front area infront of the house
#we'll fill the obj types with None since they're categorical values (non numeric)

non_numeric_cols = ['MasVnrType', 'FireplaceQu','GarageFinish', 'GarageQual','GarageCond', 'GarageType', 'BsmtExposure', 'BsmtFinType2','BsmtCond', 'BsmtFinType1', 'BsmtQual']
numeric_cols = ['GarageYrBlt', 'MasVnrArea']
#frontage_col = ['LotFrontage']
for c in non_numeric_cols:
    df[c] = df[c].fillna('None')

for c in numeric_cols:
    df[c] = df[c].fillna(0)

#gonna group it by neigborhood
df['LotFrontage'] = df.groupby('Neighborhood')['LotFrontage'].transform('median')
df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].median())   #in case there are NaNs

```

```python
df['Electrical'].value_counts(dropna=False)
```

```python
# since electrical only has 1 NaN val, we should prolly fill it with the mode instead of None
df['Electrical'] = df['Electrical'].fillna(df['Electrical'].mode()[0])
```

```python
df.isnull().sum().sort_values(ascending=False).head(16)
```

# Outlier Detection
We only need to do it on numeric cols

```python
sns.boxplot(df)   #just to check
```

```python
numeric_df = df.select_dtypes(include=['number'])
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75 )
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
filtered_df = df[~((numeric_df < lower_limit) | (numeric_df > upper_limit)).any(axis=1)]
filtered_df
```

#### DOING IT COLUMN-WISE

```python

numeric_df = df.select_dtypes(include=['number'])
clean_numeric_df = numeric_df.copy()

for col in clean_numeric_df.columns:
    Q1 = clean_numeric_df[col].quantile(0.25)
    Q3 = clean_numeric_df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    clean_numeric_df[col] = clean_numeric_df[col].where((clean_numeric_df[col] >= lower) & (clean_numeric_df[col] <= upper))
clean_numeric_df
```

```python
clean_numeric_df.isnull().sum().sort_values(ascending=False).head(33)
                                                
```
